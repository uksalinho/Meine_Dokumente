###Lists

Koleksiyonlar coklu veri tipi(multiple values) tutar. int, str gibi veriler ise single value'dir.
List'ler degitirilebilir (mutable) tiptir. Sirali elemanlar dizisidir. Listler farkli tip elemanlardan olusabilir. Listeler olusturulurken [ ] kullanilir yada list( ) seklinde döngüye alinabilir.  



```python
L = []  # boş bir liste
L0 = list() # boş bir liste
L1 = list("happy") # list sarmalina alinmis karakter listesi
L2 = ['h', 'a', 'p', 'p', 'y'] # karakter listesi, L1 ile aynı ama [list sarmalinda degil, köseli parantez icinde] 
L3 = ["happy", "clarusway", "Eric"]  # string listesi
L4 = [1, 3, 5, 7, 9, 11, 13] # integer listesi
L5 = ['a', 1, 3.145, True, [1,2,3], "Nar"]  # listeler farklı veri tiplerinden oluşabilir
nums = range(1,10)
L6 = list(nums)    # bir fonksiyon tarafından üretilen değerlerden oluşabilir
print(L, L0, L1, L2, L3, L4, L5, L6, sep="\n")
```
