Boolean Logic Expression

***and***: tüm ifadeler **True** ise bakip en sondaki **True** ifadeyi döndürür. Tüm ifadeler **False** ise bakip ilk **False** ifadeyi döndürür. 


***or*** : Hepsi **True** ise ilk **True**'yu döndürür. Hepsi **False** ise en sondaki **False**'u döndürür.

*Öncelik siralamasi* : 1. **not**, 2. **and**, 3. **or**

```python
print(True and False or not False or False )  # Önce NOT'a bakti sonra sirayla
```

None ifadeleri False olarak alir => (), [], {} ve 
0, 0.0, 0j bunlar da False. Bunlar disinda her sey True

```python
print(3 and "Ben ögrenciyim")  # and olunca ikisi de True ifade ve sonuncuyu döndü
print(3 or "Ben ögrenciyim")  # or olunca ilk True'yu ekrana basti. Hepsi False olsa en sondaki False'u basar
```
