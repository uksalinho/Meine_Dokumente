```python
# 2b. Bir sözlüğü listeye dönüştürme (dönüstürünce default olarak sadece key degerlerini verir)
d = {1: 'one', 2: 'two', 3: 'three', 4: 'four', 5: 'five'}
liste = list(d)
print(liste)
```

```python
# 2b. Bir sözlüğü listeye dönüştürme (items dersem demet olar key ve value verir)
d = {1: 'one', 2: 'two', 3: 'three', 4: 'four', 5: 'five'}
liste = list(d.items())
print(liste)
```

```python
# Sözlük yapısını güncelleme, kayıt ekleme
family = {'name1': 'Joseph',
         'name2': 'Bella',
         'name3': 'Aisha',
         'name4': 'Tom'  # Yine name4 ekledim son ekleneni alir 
        }
family.update({'name4':'Hakan', 'name5':'Ala'})
print(family)
```

```python
# d1 sözlüğünün kopyasını oluşturma seçenekleri: copy()
d1 = {'foo': 100, 'bar': 200, 'baz': 300, 'kar':400}

d2 = d1.copy()

d3 = {}  # önce d3 diye tanimladim sonra kopyaladim
d3.update(d1)

d4 = dict(d1)

d5 = dict(d1.items())

d6 = d1  # bu bir kopyalama degildir, atamadir. d1 silince d6 da silinir.
d1 == d2 == d3 == d4 == d5 == d6
print(d6)
d1.clear()  
print(d6)  # üsteki aciklamadan dolayi bos küme oldu. = ile kopyalama olmaz
print(d5)
```

```python
# Sözlüklerde arama: in, not in, get(), setdefault(),...
family = {'name1': 'Joseph',
          'name2': 'Bella',
          'name3': 'Aisha',
          'name4': 'Tom'
        }
# Aisha var mı? name4 var mı?
print('Aisha' in family)  # family.items() de yazsam ayni anlam
print('Aisha' in family.values()) ## değeri işaret ediyor
print('Aisha' in family['name3']) ## değeri işaret ediyor

print('name4' in family )
print('name4' in family.keys() )
print('name4' in family.items() )
print(family.items())  # alttaki ve üstekine False döndü. items icine girip arama yapamiyor
print('Aisha' in family.items())
```

```python
# Tüm sözlüğü silme: clear()
family = {'name1': 'Joseph', 
          'name2': 'Bella',
          'name3': 'Aisha',
          'name4': 'Tom'
         }
print(family.clear())
```

```python
# Sözlük yapısında değeri(value) dışarıdan girilen bir item'ı nasıl silebiliriz?

d = {'foo': 100, 'bar': 200, 'baz': 300, 'kar':400}

sil_value = int(input("silmek istediğiniz değeri giriniz..:"))
key_list = list(d.keys())
value_list = list(d.values())
indis = value_list.index(sil_value)
d.pop(key_list[indis])
print(d)
```

SET - KÜMELER

*   {} dict gibi süslü paratez kullanilir ama bos küme icin böyle yazilmaz 
*   ya da 
set() seklinde olusturulur


*   küme icindeki elemanlar sirasizdir
*   tekrarli elemanlari basmaz





```python
bos_kume = set()
print(bos_kume)

bos_kume_mi = {}  # set/küme degil dict/sözlük !!!
print(type(bos_kume_mi))
```

```python
set_1 = {'red', 'blue', 'yellow', 'red'}
colors = 'red', 'blue', 'yellow', 'red'
set_2 = set(colors)  # set'e cevirdi. 2 tane red var 1 tanesini basti
print(set_1)  # sira önemsiz rastgele basar
print(set_2)
print(type(colors))  # set olmadan önce tuple
```

```python
a = set('philadelphia')
b = set('dolphin')
print(a)  # küme tekrarlayanlari basmaz
print(b)  # küme tekrarlayanlari basmaz

print(a.difference(b)) # yada print(a-b) a fark b yi verir

print(a.union(b))  # yada print(a|b) birleşimi verir

print(a.intersection(b))  # yada print(a&b) kesişimi verir

# print(a+b) hata aldık
```
