```python
"""
Given a string (clarusway) and an index number int n (n=3), 
return a new string where the character at index n has been removed.
"""
text = input("text :")  # text = "clarusway"
                        # indis   012345678
n = int(input('n :'))   # n=5
new_text = text[:n] + text[n+1:]
         # 'claru' +   'way'
print(text)
print(n)
print("Output :", new_text)
```

```python
""" % operatörün işlevi
%d: int
%f: float
%s: string
%a: ascii
"""
text = "%d little pigs come out, or I'll %s, and I'll %s" % (3, 'huff', 'puff')
print(text)  # integer

text = "%f little pigs come out, or I'll %s, and I'll %s" % (3, 'huff', 'puff')
print(text)  # float

text = "%.2f little pigs come out, or I'll %s, and I'll %s" % (3, 'huff', 'puff')
print(text)  # float deyip . dan sonra 2 karakter istedim
```

```python
text = "{} little pigs come out, or I'll {}, and I'll {}".format(3, 'huff', 'puff')
print(text)  # Üstekini burda format ile %'siz olarak yazdim. 
```

```python
text = f"{3} little pigs come out, or I'll {'huff'}, and I'll {'puff'}"
print(text)  # Bu da üsteki ikisinin farkli bir sekli. f basa geldi
```

```python
num = 3.145678
print("PI Sayısı :%.2f" %num)  # % ile . dan sonra 2 karakter aldi

print("PI Sayısı :{:.2f}".format(num))  # % olmadan format metoduyla . dan sonra 2 karakter aldi. 

print("PI Sayısı :{:05.2f}".format(num))  # Yine format metodu ile 5. indise kadar al dedim

print(f"PI Sayısı : {num:.2f}")  # Format metodunun farkli gösterimi

print(f"PI Sayısı : {num:05.2f}")  # Format metodunun farkli gösterimi ile 5. indise kadar al dedim
```
