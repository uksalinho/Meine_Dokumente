**Slicing a list**

The formula syntax => list_name = [start:stop:step]

```python
liste = list(range(11))  # range sıralı/ardışık sayı dizisi oluşturur
print(liste)
print(liste[1:11:2])

liste1 = list(range(1,11,2)) # range ile artim miktari vererek ardışık tek sayı listesi
print(liste1)
```

```python
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat"]
print(liste[3:])  # 3. den basla sona kadar
print(liste[:3])  # bastan basla 3. ye kadar ( 3 dahil degil)
print(liste[::2])  # bastan basla sona 2ser atlayarak git
```

```python
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
print(liste[::-1])  # listeyi tersten yazdir
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
liste.reverse()  # revers listeyi tersten yazar
print(liste)
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
print(liste[::-2])  # - olunca tersten/sondan siraliyor
print(liste[:-6:-2])  # baslangic icin sondan siraladi
```

```python
liste = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(liste[7:3:-1]) # 7den 3e azalarak gidiyor. ters sırada 7 den 3.elemana(dahil değil) dilimler. 
print(liste[2:6:-1])  # 2den 6ya azalarak gidemez. bir şey alamıyor [] boş liste döndürür. 
```

```python
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
print(liste[0:-6:-2])
print(liste[0:1:-2])
print(liste[0:-6:2])
print(liste[0:1:2])
print(liste[-1:-6:-2])
```

```python
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
print(liste[6:1:-2])  # artım miktarı(step) negatifse sağdan sola doğru ilerleme oluyor
liste = ["elephant", "fox", "wolf", "giraffe", "rabbit", "cat", "dog"]
print(liste[6:1:1])  #artım miktarı(step) pozitifse soldan sağa doğru ilerleme oluyor
```

### Tuples

Degistirilemez sirali veri tipidir. Normal parantez kullanilir. Daha hizli calisir. Icerigine dokunulmasi istenmeyen veriler icin kullanilabilir. Sözlük yapilarinda 'key' deger olarak kullanilabilir. () yada tuple() seklinde tanimlanabilir. Parantez olmaksizin ',' ile ayrilan birden fazla degeri tuple olarak algilar.

```python
# Tuple tanımla şekilleri
t0 = ()  # boş bir demet
t1 = tuple()
t2 = 'p', 'y', 't', 'h', 'o', 'n'    # tuple=> ',' ile ayrilmis parantezsiz
t3 = ('p', 'y', 't', 'h', 'o', 'n')  # tuple=> ',' ile ayrilmis parantez icinde
t4 = 'python'  #string => "" icinde sadece 1 eleman var
t5 = 'p'   # string => "" icinde sadece 1 eleman var
t8 = ('p')   # string => parantez icinde sadece 1 eleman var
t6 = 1, 2  # tuple => parantez yok ama virgüllerle ayrılmış birden fazla eleman
t7 = ('p', ) # tuple => t4 ve t5 den farki sonunda ',' olmasi
print(type(t4), type(t5), type(t8))  
print(t6)
print(type(t2))
print(type(t7))
```

```python
t10 = (b,)  # b'yi değişken ya da nesne olarak görüyor. yani tanimlanmaliydi!
```

```python
liste = [1,2,3,4,5,6,7]
t = tuple(liste)  # Listeyi tuple yaptim
print(t)

string = "clarusway"
t = tuple(string)  # stringi tuple yaptim
print(t)

t = 1, 3, 'a', 'way'  # Farkli veri tipleri olsa da tuple
print(t)

t = 1, 3, 'a', 'way' 
u = t, (4,5,6)
print(u)  # iç içe tuple olustu
```
