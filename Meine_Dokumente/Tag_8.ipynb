```python
"Hatƒ±rlama: Python Tuples"
a = (1)  ## type: int c√ºnk√º parantez icinde , yok ve tek bir eleman var
a = (1,) ## type: tuple c√ºnk√º parantez olsa da olmasa da , ile ayrilmis elemanlar var
b = 1,2,3  ## type: tuple c√ºnk√º parantezsiz olarak sadece degerler var
c = tuple([1,2,3,4]) ## type: tuple c√ºnk√º en basta tuple dedim. 
d = tuple('a')  ## type: tuple bunlar immutable yani dogrudan degistirme yapamam
print(b, c, d, sep="\n")

print(dir(tuple))   ## #tuple methods: 'count', 'index'
tuples = (1, 2, 'hi', 3)
print(tuples.index('hi'))  ## 2 (index of 'hi') indis no verir
print(tuples.count('hi'))  ## 1 (count of 'hi') hi ifadesi kac kere geciyoru verir

print(tuples[1])    ## 1 (accessing a single item) tuple'lara b√∂yle erisiriz
# tuples[2] = 'bye' ## NO, tuples cannot be changed. Tuple'a dogrudan atama yapamam
# del tuples[2]     ## NO, tuples cannot be changed. ve yine dogrudan silemem

tuples = (1, 2, 'bye', 3 )  ## this works because new tuple
print(tuples)
a,b,c,d = tuples    ## unpacking
print(c)            ## 'bye'

tuples1 = (1,2,9,4,5,6,7,0)
print(len(tuples1)) ## 8 uzunluk
print(max(tuples1)) ## 9 en b√ºy√ºk
print(min(tuples1)) ## 0 en k√ºc√ºk
print(sum(tuples1)) ## 34 (sum of items) toplam
```

# **S√∂zl√ºk**
Hem degisken ifadesi hem de degerler tirnak icine yazilir yani key ve value! Parantez olarak {} kullanilir. {'key'='value'} buna item denir.

Sadece {} yazilabilir ya da dict() d√∂ng√ºs√ºne alinarak s√∂zl√ºk yapisi olusturulur. dict sarmalina alininca key icin "" kalkar. Dogrudan degistirme, ekleme yapilabilir. Icinde farkli veri tipleri barindirabilir.

```python
grocer1 = {'fruit':'apple', 'drink':'water'} 
grocer2 = dict(fruit='apple', drink='water')  ## E≈üdeƒüeri
print(grocer1, grocer2, sep='\n')
```

```python
"""zip function: Her tipten koleksiyon verilerini birle≈ütirerek,
bir iterasyon yapƒ±labilecek liste, demet veya s√∂zl√ºk yapƒ±sƒ± olu≈üturmada kullanƒ±lƒ±r.

Program-1: ƒ∞ki farklƒ± listeyi biri anahtar (key), diƒüeri deƒüeri (value) olacak
≈üekilde s√∂zl√ºk yapƒ±sƒ±na d√∂n√º≈üt√ºrmek i√ßin zip() fonksiyonu kullanƒ±labilir.
"""
ad = ['one','two','three','four','five']
no = [1, 2, 3, 4, 5 ]
szlk = dict(zip (ad,no))
liste = list(zip(ad,no))
print("S√∂zl√ºk: ", szlk)  # parantez {}
print("Liste: ", liste)  # parantez []
print(zip(ad,no))  # bir koleksiyon tipine d√∂n√ºst√ºrmeli
```
